```python
import numpy as np
import math
from statistics import mean
import matplotlib.pyplot as plt

def polygon(n):
    points = np.random.rand(n,2)
    for i in range(n-1):
        for j in range(i+1,n):
            if points[i].all()==points[j].all():
                points[j] = np.random.rand(1,2)
    # reordering vertices
    print(points)
    x_c = points[:,0]
    y_c = points[:,1]
    mean_x = mean(x_c)
    mean_y = mean(y_c)
    angles = []
    for i in range(len(points)):
        angle = math.atan2( (y_c[i]-mean_y),(x_c[i]-mean_x) )
        angles.append(angle)
    sort_index = np.argsort(np.array(angles))
    new_points = points[sort_index]
    return new_points
P = polygon(200)
```

```python
plt.plot([i[0] for i in P],[i[1] for i in P],'-o')
```
